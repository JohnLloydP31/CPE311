```python
def TOH(n, Sour, Aux, Des):
  if n == 1:
    print("Move Disk", n, "form", Sour, "to", Des)
  else:
    TOH(n-1,Sour, Des, Aux)
    print("Move Disk", n, "from", Sour, "to", Des)
    TOH(n-1,Sour, Des, Aux)

n = 3
TOH(n, 'A', 'B', 'C')


```
