#Exercise 1

```python
import pandas as pd

df_fb = pd.read_csv('/content/fb.csv')
df_fb.head()
```

```python
df_goog = pd.read_csv('/content/goog.csv')
df_goog.head()
```

```python
df_nflx = pd.read_csv('/content/nflx.csv')
df_nflx.head()
```

```python
df_amzn = pd.read_csv('/content/amzn.csv')
df_amzn.head()
```

```python
df_aapl = pd.read_csv('/content/aapl.csv')
df_aapl
```

```python
# assign ticker to APPLE
new_df_aapl = df_aapl.assign(
    ticker= 'AAPL'
)
new_df_aapl
```

```python
# assign ticker to AMAZON
new_df_amzn = df_amzn.assign(
    ticker='AMZN'
)
new_df_amzn
```

```python
# assign ticker to FACEBOOK
new_df_fb = df_fb.assign(
    ticker='FB'
)
new_df_fb
```

```python
# assign ticker to GOOGLE
new_df_goog  = df_goog.assign(
    ticker='GOOG'
)
new_df_goog
```

```python
# assign ticker to NETFLIX
new_df_nflx = df_nflx.assign(
    ticker = 'NFLX'
)
new_df_nflx
```

```python
# append dataframes into a single dataframe
df_append = new_df_aapl.append([new_df_amzn, new_df_fb, new_df_goog, new_df_nflx], ignore_index=True)
df_append
```

```python
# save the result into a new csv file
df_append.to_csv('faang.csv', index=False)
```

#Exercise 2

```python
# read faang.csv
df_faang = pd.read_csv('/content/faang.csv')
df_faang
```

```python
# converting date to datetime and volume to integer
df_faang.loc[:,'date'] = pd.to_datetime(df_faang.date)
df_faang.loc[:, 'volume'] = df_faang.loc[:, 'volume'].astype(int)
df_faang.dtypes
```

```python
# 7 highest volume
largest_volume = df_faang.nlargest(n=7, columns = 'volume')
largest_volume
```

```python
# using melt() operation
melt_faang = df_faang.melt(
    id_vars= ['date', 'ticker']
)
melt_faang
```
