#Mean

```python
salaries_sum = sum(salaries)
mean = salaries_sum / len(salaries)
print("Mean salary: ", mean)
```

# Using statistics module

```python
import statistics as st
mean_salaries = st.mean(salaries)
print("Mean salary: ", mean_salaries)
```

#Median

```python
sorted_salaries = sorted(salaries)
n = len(sorted_salaries)

if n % 2 == 0:
  median = (sorted_salaries[n//2] + sorted_salaries[n//2 - 1]) / 2
else:
  median = (sorted_salaries[n//2])
print("Median salary: ", median)
```

# Using statistics module

```python
import statistics as st
median_salaries = st.median(sorted_salaries)
print("Median salary: ", median_salaries)
```

#Mode

```python
from collections import Counter
def mode_salaries(salary):
  count_salaries = Counter(salary)
  max_count = max(count_salaries.values())

  mode_salaries = [salary for salary, count in count_salaries.items() if count == max_count]
  return mode_salaries
mode = mode_salaries(salaries)
print("Mode salary:", mode)
```

```python
import statistics as st
mode_salaries = st.mode(salaries)
print("Mode salary: ", mode_salaries)
```

#Sample Variance

```python
n = len(salaries)
x = sum(salaries) / n
sum_deviations = 0

for salary in salaries:
    deviation = (salary - x) ** 2
    sum_deviations += deviation

variance = sum_deviations / (n - 1)
print("Variance salary:", variance)
```

```python
import statistics as st
variance_salaries = st.variance(salaries)
print("Variance salary: ", variance_salaries)
```

# Sample standard deviation

```python
sd = variance ** .5
print("Standard deviation salary: ", sd)
```

```python
import statistics as st
sd_salaries = st.stdev(salaries)
print("Standard deviation salary: ", sd_salaries)
```

Using the same data, calculate the following statistics using the functions in the statistics module where appropriate:

- Coefficient of variation Interquartile range

#Range

```python
min_salary = min(salaries)
max_salary = max(salaries)

range = max_salary - min_salary
print("Range salary: ", range)
```

#Coefficient of variation

```python
import statistics as st
CoV = st.stdev(salaries) / st.mean(salaries) * 100
print("Coefficient of variation salary: ", CoV)
```

 # Interquartile range

```python
import statistics as st
def IR(salaries):
  q1, q2, q3 = st.quantiles(salaries, n = 4)
  ir = q3 - q1
  return ir

ir = IR(salaries)
print("Interqurtile range: ", ir)
```

# Quartile coefficient of dispersion

```python
import statistics as st
def QCoD(salaries):
  q1, q2, q3, = st.quantiles(salaries, n =4)
  qcd = (q3 - q1) / (q3 + q1)
  return qcd

qdc = QCoD(salaries)
print("Quartile coefficient of dispersion salary: ", qdc)

```
